```python
aapl = pd.read_csv('aapl.csv')
amzn = pd.read_csv('amzn.csv')
fb = pd.read_csv('fb.csv')
goog = pd.read_csv('goog.csv')
nflx = pd.read_csv('nflx.csv')
aapl['ticker'] = 'AAPL'
amzn['ticker'] = 'AMZN'
fb['ticker'] = 'FB'
goog['ticker'] = 'GOOG'
nflx['ticker'] = 'NFLX'
```

```python
faang = pd.concat([aapl,amzn,fb,goog,nflx])
faang
```

```python
faang.to_csv('faang.csv', index=False)
```

```python
faang = faang.assign(
    date = lambda x: pd.to_datetime(x.date),
    volume = lambda x: x.volume.astype('int'))
```

```python
faang.dtypes
```

```python
faang = faang.sort_values(by = ['date', 'ticker'], ascending=[True, True])
faang.head(10)
```

```python
lowest_vol = faang.sort_values(by=['volume'], ascending = True)
lowest_vol.head(7)
```

Covid Dataset

```python
covid19 = pd.read_csv('covid19_cases.csv')
covid19.head()
```

```python
covid19['date'] = pd.to_datetime(covid19.dateRep)
covid19.dtypes
```

```python
covid19.set_index('date', inplace=True)
```

```python
covid19.sort_index()
covid19.head()
```

```python
covid19 = covid19.replace({'United_States_of_America' : 'USA', 'United_Kingdom': 'UK'})
```

```python
covid19_new = covid19.loc[covid19['countriesAndTerritories'].isin(['Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy', 'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey', 'USA', 'UK'])]
covid19_new.head()
```

```python
covid19_new = covid19_new.fillna(0)
pivot_covid = covid19_new.pivot(columns='countriesAndTerritories', values='cases').fillna(0)
pivot_covid.head()
```
